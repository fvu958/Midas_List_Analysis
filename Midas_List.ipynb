```python
education_others = []

for name in others:
    response = requests.get(url+name)
    soup = bs(response.content, 'html.parser')
    span = soup.findAll('span', {'class':'profile-list__item__text'})
    
    try:
        edu = span[-1].text
    except Exception:
        print('N/A')
    education_others.append(edu)
```

```python
data = [('Name', midas18),
       ('Education', education)]


df = pd.DataFrame.from_items(data)
df.head()
```

```python
data_others = [('Name', others),
       ('Education', education_others)]


df_others = pd.DataFrame.from_items(data_others)
df_others.head()
```

```python
df = pd.concat([df, df_others], ignore_index=True)
df
```

```python
names = df['Name'].tolist()
clean_names = []

for name in names:
    new = name.replace('-', ' ')
    capital = new.title()
    clean_names.append(capital)

clean_names[:5]
```
